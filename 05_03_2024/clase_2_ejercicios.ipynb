#### ¡Hola! ¡Necesitamos tu ayuda! Tenemos un pequeño problema en nuestro sistema de citas médicas. Sucede que al momento de registrar los datos del usuario, se nos olvidó implementar una validación para poder manejar los números telefónicos en un formato específico. En este caso, tenemos lo siguiente:


numero_telefonico = "+52 123 456 78 90"

#### Entonces, no se pueden realizar correctamente las llamadas debido a que nuestro sistema ya incluye por defecto el prefijo +52, lo que duplica el valor, por ejemplo: +52 +52 555 555 55 55. Por lo tanto, necesitamos que filtres los datos y elimines el +52 que está de más en la lista.

#### Ademas, también necesito que elimines los espacios demas que existen al inicio y al final de los numeros, ya que estos caracteres vacios, ocupan información innecesaria en nuestra base de datos.

#### ¡Suerte!



```python
numero_telefonico = "   +521234567890       "
numero_telefonico_con_prefijo = "+52 " + numero_telefonico

numero_telefonico_con_prefijo = numero_telefonico_con_prefijo.removeprefix('+52')
numero_telefonico_con_prefijo = numero_telefonico_con_prefijo.strip()

print(numero_telefonico_con_prefijo)
```

### Ejercicio 2

#### Oye, tengo que hacer próximamente un examen de geometría y necesito repasar varios temas, entre ellos el cálculo de área de las siguientes figuras:



#### ¿Será que me podrías crear una pequeña calculadora la cual almacene los datos en una variable, despues los asigne a una formula y luego imprima su resultado? Esto me ayudaría mucho para corroborar que mis datos son correctos. ¡Saludos!




```python
# Fórmula para calcular el área de un cuadrado
lado = 5
area_cuadrado = lado * lado

# Fórmula para calcular el área de un rectángulo
base = 10
altura = 7
area_rectangulo = base * altura

# Fórmula para calcular el área de un triángulo
base = 6
altura = 8
area_triangulo = (base * altura) / 2

# Fórmula para calcular el área de un paralelogramo
base = 12
altura = 5
area_paralelogramo = base * altura

# Fórmula para calcular el área de un trapecio
base_mayor = 15
base_menor = 10
altura = 8
area_trapecio = ((base_mayor + base_menor) * altura) / 2

# Fórmula para calcular el área de un círculo
radio = 4
import math
area_circulo = math.pi * radio ** 2

# Fórmula para calcular el área de un rombo
diagonal_mayor = 10
diagonal_menor = 8
area_rombo = (diagonal_mayor * diagonal_menor) / 2

# Fórmula para calcular el área de un polígono regular
perimetro = 40
apotema = 6
area_poligono_regular = (perimetro * apotema) / 2

# Imprimir los resultados
print("Área del Cuadrado:", area_cuadrado)
print("Área del rectángulo:", area_rectangulo)
print("Área del triángulo:", area_triangulo)
print("Área del Paralelogramo:", area_paralelogramo)
print("Área del Trapecio:", area_trapecio)
print("Área del Círculo:", area_circulo)
print("Área del Rombo:", area_rombo)
print("Área del Polígono Regular:", area_poligono_regular)
```
